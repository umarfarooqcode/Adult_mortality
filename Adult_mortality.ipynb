```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
sns.set(style="whitegrid")

# Distribution of Adult Mortality Rates (AMR) for Males and Females
plt.figure(figsize=(14, 6))
sns.histplot(Adult_mortality_df['AMR_male(per_1000_male_adults)'], color="skyblue", label="Male", kde=True, binwidth=10)
sns.histplot(Adult_mortality_df['AMR_female(per_1000_female_adults)'], color="red", label="Female", kde=True, binwidth=10)
plt.title('Distribution of Adult Mortality Rates for Males and Females')
plt.xlabel('Mortality Rate per 1000 Adults')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# GDP per Capita vs. Health Expenditure
plt.figure(figsize=(14, 6))
sns.scatterplot(data=Adult_mortality_df, x='Average_GDP_per_capita($)', y='Average_HEXP($)', hue='Continent')
plt.title('GDP per Capita vs. Health Expenditure')
plt.xlabel('GDP per Capita ($)')
plt.ylabel('Health Expenditure ($)')
plt.legend()
plt.show()

# Average Crude Death Rate (CDR) by Continent
plt.figure(figsize=(14, 6))
sns.barplot(x='Continent', y='Average_CDR', data=Adult_mortality_df, ci=None)
plt.title('Average Crude Death Rate by Continent')
plt.xlabel('Continent')
plt.ylabel('Average Crude Death Rate')
plt.show()

# Development Level and Adult Mortality Rates
plt.figure(figsize=(14, 6))
sns.boxplot(x='Development_level', y='AMR_male(per_1000_male_adults)', data=Adult_mortality_df)
plt.title('Development Level and Male Adult Mortality Rates')
plt.xlabel('Development Level')
plt.ylabel('Male Mortality Rate per 1000 Male Adults')
plt.show()
```
